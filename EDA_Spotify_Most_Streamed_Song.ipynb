```python
import pandas as pd

def generate_metadata(dataframe):
    """
    Gera um dataframe contendo metadados das colunas do dataframe fornecido.

    :param dataframe: Dataframe para o qual os metadados serão gerados.
    :return: Dataframe contendo metadados.
    """

    # Coleta de metadados básicos
    metadata = pd.DataFrame({
        'nome_variavel': dataframe.columns,
        'tipo': dataframe.dtypes,
        'qt_nulos': dataframe.isnull().sum(),
        'percent_nulos': round((dataframe.isnull().sum() / len(dataframe)) * 100, 2),
        'cardinalidade': dataframe.nunique(),
    })

    metadata = metadata.sort_values(by=['percent_nulos', 'tipo', 'cardinalidade'])
    metadata = metadata.reset_index(drop=True)

    return metadata
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# filtrar apenas as colunas numéricas
df_numericas = df.select_dtypes(include=['float64', 'int64'])

# Definir o número de colunas e o número de linhas de subplots automaticamente
num_colunas = df_numericas.shape[1]
num_linhas = math.ceil(num_colunas / 3) # Definir 3 gráficos por linha

# Definir o tamanho da figura com base na quantidade de subplots
plt.figure(figsize=(15, num_linhas * 5))

# Loop para criar um histograma para cada coluna numérica
for i, coluna in enumerate(df_numericas.columns, 1):
    plt.subplot(num_linhas, 3, i) # Ajustar a grade de subplots dinamicamente
    sns.histplot(df_numericas[coluna], kde=True, bins=30) # kde=True adiciona a linha de densidade
    plt.title(f'Distribuição de {coluna}')

plt.tight_layout() # Ajusta os gráficos para não sobreporem
plt.show()
```
